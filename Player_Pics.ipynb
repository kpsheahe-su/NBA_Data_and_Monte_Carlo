```python
import string
import pandas as pd
import urllib.request
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', 500)
```

```python
file = r'C:\Users\michi\Downloads\PlayerID+PlayerName.csv'
file_df = pd.read_csv(file) 
len(file_df)
```

```python
players = file_df['PLAYER_ID']
players
```

```python
x = 0

#choose file type
filetypepng = '.png'
filetypejpg = '.jpg'
file_df['file_name'] = np.nan
file_df['player_profile_link'] = np.nan

for t,i in enumerate(players):
    playerID = str(i)
    try: 
        url = 'https://ak-static.cms.nba.com/wp-content/uploads/headshots/nba/latest/260x190/'+playerID+'.png'
        player_name = file_df.iat[t,1] 
        player_name = string.capwords(player_name)
        file_df.iat[t,1] = player_name
        file_df.iat[t,2] = player_name+filetypejpg
        file_df.iat[t,3] =  url
        urllib.request.urlretrieve(url, player_name+filetypejpg)
        print(url, player_name+filetypejpg) 
    except:
        silhouette_url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcSNcvgaCmOfXMpmKhtsbSDxYliiugxUIVhEE0zUvZcN6Zok1KRHGkB5xwHEHTp_kRoFcpk&usqp=CAU'
        urllib.request.urlretrieve(silhouette_url, player_name+filetypejpg)
        file_df.iat[t,3] =  silhouette_url
        print('Error, placing player profile pic as a silhouette', silhouette_url, player_name+filetypejpg)
        x+=1
print(x)
    
```
